```python

class Animal:

    def __init__(self,name,location,age):
        self.firstname=name
        self.thelocation=location
        self.age=age
        
        
        
    def printlocation(self):
        print(self.firstname+" lives in "+self.thelocation)
        
        
    def printage(self):
        print(self.firstname+"'s' age is: " + str(self.age))
        
    def oneyearolder(self):
        
        print("One year passes...")
        
        self.age+=1
        
    def transfertoIstanbul(self):
        self.thelocation="İstanbul"
        print(self.firstname+" is succesfuly transferred to "+self.thelocation)
       
    





class Dog(Animal):
    
    
    
    
    def bark(self):
        
        print("Woof")
    
    def walking(self):
        
        print("Dog "+(self.firstname)+" is walking in "+ self.thelocation)
    
    
    
class Cat(Animal):
    
   
        
        
    def meow(self):
        
        print("Meoww")
        
    def walking(self):
        
        
        print("Cat "+self.firstname+" is walking in "+self.thelocation)
        
        
        
        
        
        

    
    
    
    
```

```python
dog1=Dog("Maço", "Ankara",5)
dog1.walking()
dog1.bark()
dog1.printlocation()
dog1.printage()
dog1.oneyearolder()
dog1.printage()
dog1.transfertoIstanbul()
dog1.printlocation()


cat1=Cat("Garfield","New Jersey",8)
cat1.walking()
cat1.meow()
cat1.printlocation()
cat1.printage()
cat1.oneyearolder()
cat1.printage()
cat1.transfertoIstanbul()
cat1.printlocation()

cat2=Cat("Böğürtlen","Eskişehir",2)
cat2.walking()
cat2.meow()
cat2.printlocation()
cat2.printage()
cat2.oneyearolder()
cat2.printage()
cat2.transfertoIstanbul()
cat2.printlocation()
```
