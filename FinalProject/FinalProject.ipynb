```python
global emplang
global manlang
global complang

global empinfo
global maninfo
global compinfo

compinfo={}
complang=set()


empinfo={}
emplang=set()


maninfo={}
manlang=set()






class company:
    
    
        
        
    
        
        
              
    def companyinfo(self):
        print("Company information: ")
        print(compinfo)
        
    def managerinfo(self):
        
        print("Managers information: ")
        print(maninfo)
        
    def employeeinfo(self):
        
        print("Employees information: ")
        print(empinfo)
               
    
    
    def totalcomplang(self):
        print("All company can speak: "+str(complang))
    
    def totalmanlang(self):
        
        
        print("Managers can speak: "+str(manlang))
        
  
    def totalemplang(self):
        
        
        print("Employees can speak: "+str(emplang))
        




class employees:
    
    
    
    
    
    
    def __init__(self,name,surname,languages):
        
        
        self.name=name
        self.surname=surname
        self.languages=languages
        
        empinfo[self.name+" "+self.surname]=self.languages
        compinfo["Employees"]=empinfo
        
        for i in self.languages:
        
            emplang.add(i)
            complang.add(i)
        
        
      
        

        
  
    def addemplang(self,lang):
        
        print(self.name+" "+self.surname+" has learned: "+ lang)
        
        self.languages.add(lang)
        empinfo[self.name+" "+self.surname]=self.languages
        compinfo["Employees"]=empinfo
  

    def introduceself(self):
        print("Hello, I am a an employee, my name is "+self.name+" "+self.surname)
        print("And I can speak total of: ")
        for i in self.languages:
            print(str(i))
            

class managers:
    
    
    
    
    def __init__(self,name,surname,languages):
        
        
        self.name=name
        self.surname=surname
        self.languages=languages
        
        
      
        
        maninfo[self.name+" "+self.surname]=self.languages
        compinfo["Managers"]=maninfo
        
        for i in self.languages:
        
            manlang.add(i)
            complang.add(i)
        
                
        
  
    def addmanlang(lang):
        
        print(self.name+" "+self.surname+" has learned: "+ lang)
        
        self.languages.add(lang)
        maninfo[self.name+" "+self.surname]=self.languages
        compinfo["Managers"]=maninfo 

    def introduceself(self):
        print("Hello, I am a manager, my name is "+self.name+" "+self.surname)
        print("And I can speak total of: ")
        for i in self.languages:
            print(str(i))


    

        
        
#I created a parent class "company" only to print the total information way more easily.
        
```
